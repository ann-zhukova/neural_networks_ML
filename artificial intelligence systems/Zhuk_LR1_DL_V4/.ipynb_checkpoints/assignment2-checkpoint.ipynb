## Полносвязная нейронная сеть

Реализовать нейронную сеть, состоящую из двух полносвязных слоев и решающую задачу классификации на наборе данных из лабораторной работы 1.

```python
import numpy as np
import matplotlib.pyplot as plt

from scripts.classifiers.neural_net import TwoLayerNet

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) 
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


def rel_error(x, y):
    """ returns relative error """
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))
```

1. Добавьте реализации методов класса TwoLayerNet . Проверьте вашу реализацию на модельных данных (Код приведен ниже).  

```python
input_size = 4
hidden_size = 10
num_classes = 3
num_inputs = 5

def init_toy_model():
    np.random.seed(0)
    return TwoLayerNet(input_size, hidden_size, num_classes, std=1e-1)

def init_toy_data():
    np.random.seed(1)
    X = 10 * np.random.randn(num_inputs, input_size)
    y = np.array([0, 1, 2, 2, 1])
    return X, y

net = init_toy_model()
X, y = init_toy_data()
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

digits = load_digits()
```

```python
try:
   del X_train, y_train
   del X_test, y_test
   print('Clear previously loaded data.')
except:
   pass

X_train,X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=2017)

print('Training data shape: ', X_train.shape)
print('Training labels shape: ', y_train.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))
print( X_train.shape, X_test.shape, y_train.shape)
X_train,X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=2017)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=2017)
```

```python
mean_image = np.mean(X_train, axis=0)
print(mean_image[:10])

X_train -= mean_image
X_val -= mean_image
X_test -= mean_image

print(X_train.shape, X_val.shape, X_test.shape)
```

```python
input_size = 64
hidden_size = 50
num_classes = 10
net = TwoLayerNet(input_size, hidden_size, num_classes)

stats = net.train(X_train, y_train, X_val, y_val,
            num_iters=1000, batch_size=100,
            learning_rate=0.01, learning_rate_decay=0.95,
            reg=0.25, verbose=True)

val_acc = (net.predict(X_val) == y_val).mean()
print('Validation accuracy: ', val_acc)
```

Для отладки процесса обучения часто помогают графики изменения loss и accuracy в процессе обучения. Ниже приведен код построения таких графиков. 

```python
input_size = 8 * 8 * 1
hidden_size = 100
num_classes = 10
learning_rates = [0.001,0.1,0.0001]
regularization_strengths = [0, 1, 10]
num_iters = [100, 1000, 1000]
best_val = 0
for lr in learning_rates:
    for reg in regularization_strengths:
        for iters in num_iters:
            print("Learning rate",lr,"reg",reg, "Number of iteration",iters)
            net = TwoLayerNet(input_size, hidden_size, num_classes)
            stat = net.train(X_train, y_train, X_val, y_val,
            num_iters=iters, batch_size=100,
            learning_rate=lr, learning_rate_decay=0.95,
            reg=reg, verbose=True)
        
            val_acc = (net.predict(X_val) == y_val).mean()
            print(val_acc)
        
            if best_val < val_acc:
                best_val = val_acc
                best_parameters = { 'Learning rate':lr, 'reg': reg, 'Number of iteration':iters}

best_net = TwoLayerNet(input_size, hidden_size, num_classes)
stat = best_net.train(X_train, y_train, X_val, y_val,
            num_iters=best_parameters['Number of iteration'], batch_size=100,
            learning_rate=best_parameters['Learning rate'], learning_rate_decay=0.95,
            reg=best_parameters['reg'], verbose=True)
```
